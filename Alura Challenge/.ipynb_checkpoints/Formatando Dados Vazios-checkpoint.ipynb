# Formatando dados

Precisarei formatar os dados que estavam vazios para aplicar o modelo preditivo.

```python
import pandas as pd
```

```python
dataset_vazios = pd.read_csv(r'C:\Users\Jose Eduardo\Documents\GitHub\alura_challenge\Dados\dataset_vazios.csv')
dataset_vazios.head(3)
```

Teremos que fazer todo o tratamento que foi feito anteriormente também para esses dados, primeiro iremos traduzir tudo.

```python
dataset_vazios.columns = ['Identificação do Cliente', 
                          'Churn', 
                          'Sexo', 
                          'Idoso',
                          'Casado', 
                          'Dependentes', 
                          'Meses de Contrato', 
                          'Serviço Telefônico', 
                          'Multiplas Linhas', 
                          'Serviço de Internet',
                          'Proteção Internet',
                          'Backup Online',
                          'Proteção Dispositivo',
                          'Suporte Técnico',
                          'TV a Cabo',
                          'Stremaming Filmes',
                          'Tipo do Contrato',
                          'Fatura Online',
                          'Forma de Pagamento',
                          'Total Gasto Mensal',
                          'Total Gasto Anual'
                         ]

dataset_vazios.replace('No', 'Não', inplace=True)
dataset_vazios.replace('Yes', 'Sim', inplace=True)
dataset_vazios.replace('Fiber optic', 'Fibra ótica', inplace=True)
dataset_vazios.replace('Male', 'Masculino', inplace=True)
dataset_vazios.replace('Female', 'Feminino', inplace=True)
dataset_vazios.replace('No internet service', 'Não tem serviço de internet', inplace=True)
dataset_vazios.replace('Month-to-month', 'Mês a mês', inplace=True)
dataset_vazios.replace('Two year', 'Dois anos', inplace=True)
dataset_vazios.replace('One year', 'Um ano', inplace=True)
dataset_vazios.replace('Electronic check', 'Cheque eletrônico', inplace=True) 
dataset_vazios.replace('Mailed check', 'Cheque', inplace=True)
dataset_vazios.replace('Bank transfer (automatic)', 'Transferência bancária (automática)', inplace=True)
dataset_vazios.replace('Credit card (automatic)', 'Cartão de crédito (automático)', inplace=True) 
dataset_vazios.replace('No phone service', 'Não tem serviço de telefone', inplace=True)
```

```python
dataset_vazios.head(3)
```

Aplicando o Encoding nesses dados.

```python
dataset_vazios = dataset_vazios.drop(columns=['Identificação do Cliente', 'Total Gasto Anual'], axis=1)
     
dict_nao = {'Não tem serviço de internet' : 'Não',
           'Não tem serviço de telefone' : 'Não'}
dataset_vazios = dataset_vazios.replace(dict_nao)

colunas_categoricas = ['Serviço de Internet', 'Tipo do Contrato', 'Forma de Pagamento']
dataset_vazios = pd.get_dummies(data=dataset_vazios, columns=colunas_categoricas)

mapa = {'Masculino': 0, 'Feminino': 1}
dataset_vazios.Sexo = dataset_vazios.Sexo.replace(mapa)

mapa = {'Não':0, 'Sim':1}
dataset_vazios = dataset_vazios.replace(mapa)
dataset_vazios.head(3)
```

```python
dataset_vazios.to_csv(r'C:\Users\Jose Eduardo\Documents\GitHub\alura_challenge\Dados\dataset_vazios_formatado.csv', index=False)
```
