### Leitura do Arquivo JSON

```python
import pandas as pd
```

```python
dataset = pd.read_json(r'C:\Users\Jose Eduardo\Documents\GitHub\alura_challenge\Telco-Customer-Churn.json')
dataset.head()
```

Verificando o arquivo 'Telco-Customer-Churn.json' entendemos que as colunas: customer, phone, internet, account tem mais dados que precisaremos analizar, iremos separar todos esses dados para entender qual o melhor caminho tomar.

### Separando customer, phone, internet, account

#### Dados Coluna 'customer'

```python
dataset_customer = pd.json_normalize(dataset.customer)
dataset_customer.head(3)
```

#### Dados Coluna 'phone'

```python
dataset_phone = pd.json_normalize(dataset.phone)
dataset_phone.head(3)
```

#### Dados Coluna 'internet'

```python
dataset_internet = pd.json_normalize(dataset.internet)
dataset_internet.head(3)
```

#### Dados Coluna 'account'

```python
dataset_account = pd.json_normalize(dataset.account)
dataset_account.head(3)
```

#### Unindo os DataFrames

Agora que temos todas as colunas separadas, entendemos que no nosso DATASET principal, as colunas: customer, phone, internet, account tem muito mais dados que precisaremos verificar, dessa forma o melhor a se fazer é unir as colunas que separamos acima. Para fazer isso iremos criar um novo DataFrame que terá as colunas: 'customerID', 'Churn' do DataFrame original e os 4 DataFrames que separamos acima. Para isso utilizaremos o método concat do Pandas.

```python
dataset_concat = pd.concat([dataset[['customerID', 'Churn']],dataset_customer, dataset_phone, dataset_internet, dataset_account], axis=1)
dataset_concat.head(3)
```

Como estou estudando achei interessante, além de concatenar os dados com o .concat(), utilizar também o .merge() para fixar melhor o conhecimento. Agora temos um UNICO DataFrame com todos os dados que iremos análisar.

```python
dataset_merge_customer_phone = pd.merge(
    left = dataset_customer,
    right = dataset_phone,
    left_index = True,
    right_index = True
)
dataset_merge_customer_phone.head(3)
```

```python
dataset_merge_internet_account = pd.merge(
    left= dataset_internet,
    right= dataset_account,
    left_index=True,
    right_index=True
)
dataset_merge_internet_account.head(3)
```

```python
dataset_merge = pd.merge(left=dataset_merge_customer_phone, right=dataset_merge_internet_account, left_index=True, right_index=True)
dataset_merge.head(3)
```

```python
dataset_concat.head(3)
```

### Entender quais informações o conjunto de dados possui

Nessa etapa do projeto já unificamos todos os DataFrames, agora precisamos usar o [Dicionário de Dados](https://github.com/sthemonica/alura-voz/blob/main/dicionario.md) para entender do que se trata cada uma das colunas.

<img src="didionário de dados.png">

Análisando o didionário de dados entendemos o que cada uma das colunas do nosso dataset se tratam, com isso podemos ir para o próximo passo que será verificar quais os TIPOS DE DADOS do nosso dataset.

### Analisar quais os tipos de dados

Para iniciarmos a análise dos tipos de dados usaremos a função .info() em nosso DataFrame.

```python
dataset_concat.info()
```

Analisando as informações recebidas entendemos que nenhuma das colunas tem dados 'NULOS', na coluna SeniorCitizen está como inteiro, a coluna tenure está como inteiro, a coluna Charges.Monthly está como float. Analisando os dados da coluna Charges.Total verificamos que essa coluna se trata do TOTAL GASTO PELO CLIENTE, ou seja Charges.Monthly * tenure, logo encontramos uma falha, pois essa coluna também deveria ser uma coluna do tipo float. Agora precisamos identificar possiveis inconcistencias nos dados de cada coluna.

### Verificar quais são as inconsistências nos dados

Para verificar possiveis inconsistencias faremos um loop para verificar todas as 21 colunas de nosso DataFrame conforme abaixo:

```python
for data in dataset_concat:
    print('='*50)
    print()
    print(dataset_concat[data].value_counts())
    print()
```

Com isso identificamos que na coluna Churn existem 224 registros VAZIOS, diferente de todas as colunas do tipo bool (Yes ou No) a coluna SeniorCitizen está com 0 para Não e 1 para Sim, na coluna tenure existem 11 registros como 0, na coluna Charges.Total existem 11 registros VAZIOS. 

### Corrigir as inconsistências nos dados

Conforme erros identificados acima devemos corrigi-los. Corrigiremos primeiro a coluna Churn, como o intuito dessa analise é criar a persona para o Churn iremos excluir os 224 registros vazios do nosso dataset, para isso utilizaremos a função .drop()

```python
vazios = dataset_concat[dataset_concat['Churn'] == ''].index
dataset_concat.drop(vazios, inplace=True)
```

```python
dataset_concat['Churn'].info()
```

Com isso excluimos os 224 registros da nossa coluna Churn, agora precisamos resetar o Index do dataset, para isso utilizaremos a função .reset_index()

```python
dataset_concat.reset_index(inplace=True)
```

```python
dataset_concat.drop(["index"], axis=1, inplace=True)
```

Agora a inconsistência na coluna Churn foi resolvida. O próximo passo é alterar os valores da coluna SeniorCitizen, para isso utilizaremos a função .replace()

```python
dataset_concat['SeniorCitizen'].replace(0, 'No', inplace=True)
dataset_concat['SeniorCitizen'].replace(1, 'Yes', inplace=True)
```

```python
dataset_concat['SeniorCitizen'].value_counts()
```

Dessa forma a inconsistência na coluna SeniorCitizen foi resolvida.

Na coluna Charges.Total existem 11 registros VAZIOS, isso não faz sentido, pois na coluna Charges.Monthly os 11 registros estão preenchidos, o que devemos fazer é repetir o valor da coluna Charges.Monthly na coluna Charges.Total, para isso criaremos uma variavel do tipo INDEX e utilizaremos a função .loc() para acessar o grupo de linhas que tem o valor VAZIO e após isso utilizaremos a função .loc() para acessar o grupo de linhas de Charges.Total e atribuiremos o valor.

```python
idx = dataset_concat[dataset_concat['Charges.Total'] == ' '].index
dataset_concat.loc[idx, "Charges.Total"] = dataset_concat.loc[idx, "Charges.Monthly"]
```

Agora se usarmos a função .query() para fazer uma consulta na coluna Charges.Total, Charges.Monthly e tenure conseguiremos verificar que os dados são iguais.

```python
dataset_concat.query('tenure == 0')[['Charges.Total', 'Charges.Monthly', 'tenure']]
```

Conforme DataFrame acima foi resolvido a inconsistência na coluna Charges.Total. Agora podemos resolver a ultima inconsistêmcia que é o TIPO da coluna Charges.Total, iremos transformala em float com a função .astype(float).

```python
dataset_concat['Charges.Total'] = dataset_concat['Charges.Total'].astype(float)
```

```python
dataset_concat.info()
```

Conforme acima a coluna Charges.Total agora também é do tipo float que é o correto para essa coluna.

### Traduzir Colunas

Para melhor entendimento, foi nos solicitado que traduzissemos as colunas para PT-BR


```python
dataset_concat.columns = ['Identificação do Cliente', 
                          'Churn', 
                          'Sexo', 
                          'Idoso',
                          'Casado', 
                          'Dependentes', 
                          'Meses de Contrato', 
                          'Serviço Telefônico', 
                          'Multiplas Linhas', 
                          'Serviço de Internet',
                          'Proteção Internet',
                          'Backup Online',
                          'Proteção Dispositivo',
                          'Suporte Técnico',
                          'TV a Cabo',
                          'Stremaming Filmes',
                          'Tipo do Contrato',
                          'Fatura Online',
                          'Forma de Pagamento',
                          'Total Gasto Mensal',
                          'Total Gasto Anual'
                         ]
dataset_concat.head(3)
```

O próximo passo é traduzir todo o conteúdo do DataFrame, usaremos o .replace() para isso.

```python
dataset_concat.replace('No', 'Não', inplace=True)
dataset_concat.replace('Yes', 'Sim', inplace=True)
dataset_concat.replace('Fiber optic', 'Fibra ótica', inplace=True)
dataset_concat.replace('Male', 'Masculino', inplace=True)
dataset_concat.replace('Female', 'Feminino', inplace=True)
dataset_concat.replace('No internet service', 'Não tem serviço de internet', inplace=True)
dataset_concat.replace('Month-to-month', 'Mês a mês', inplace=True)
dataset_concat.replace('Two year', 'Dois anos', inplace=True)
dataset_concat.replace('One year', 'Um ano', inplace=True)
dataset_concat.replace('Electronic check', 'Cheque eletrônico', inplace=True) 
dataset_concat.replace('Mailed check', 'Cheque', inplace=True)
dataset_concat.replace('Bank transfer (automatic)', 'Transferência bancária (automática)', inplace=True)
dataset_concat.replace('Credit card (automatic)', 'Cartão de crédito (automático)', inplace=True) 
dataset_concat.head(3)
```

Com isso todo o conteúdo do DataFrame foi traduzido para Português Brasil.
