### Leitura do Arquivo JSON

Primeiramente iremos ler o arquivo JSON disponibilizado pela API da Alura Voz e entender todos os dados. Para isso utilizaremos o Pandas.

```python
import pandas as pd
```

```python
dataset = pd.read_json('Telco-Customer-Churn.json')
dataset.head()
```

Verificando o arquivo 'Telco-Customer-Churn.json' entendemos que as colunas: customer, phone, internet, account tem mais dados que precisaremos analizar, iremos separar todos esses dados para entender qual o melhor caminho tomar.

### Separando customer, phone, internet, account

```python
dataset_customer = pd.json_normalize(dataset.customer)
dataset_customer
```

```python
dataset_phone = pd.json_normalize(dataset.phone)
dataset_phone
```

```python
dataset_internet = pd.json_normalize(dataset.internet)
dataset_internet
```

```python
dataset_account = pd.json_normalize(dataset.account)
dataset_account
```
